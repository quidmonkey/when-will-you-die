```python
import pandas as pd

deaths_csv_url = "https://data.delaware.gov/api/views/nck5-dhqv/rows.csv?accessType=DOWNLOAD"
deaths_df = pd.read_csv(deaths_csv_url)
```

```python
target_columns = ["COUNTY OF RESIDENCE", "GENDER", "RACE", "AGE GROUP", "EDUCATION", "CAUSE OF DEATH", "PLACE OF DEATH"]
ddf = deaths_df[target_columns]
ddf.dropna()
ddf.count()
```

```python
ddf.groupby("COUNTY OF RESIDENCE").count()
```

```python
# remove out of state education
ddf1 = ddf[ddf["COUNTY OF RESIDENCE"] != "Out of State"]
ddf1.groupby("COUNTY OF RESIDENCE").count()
```

```python
ddf1.groupby("GENDER").count()
```

```python
ddf1.groupby("RACE").count()
```

```python
# remove unknown races
unknowns = ["UNKNOWN", "Two or more races"]
ddf2 = ddf1[~ddf1.RACE.isin(unknowns)]
ddf2.groupby("RACE").count()
```

```python
ddf2.groupby("EDUCATION").count()
```

```python
# remove unknown education
ddf3 = ddf2[ddf2.EDUCATION != "Unknown"]
ddf3.groupby("EDUCATION").count()
```

```python
ddf3.groupby("AGE GROUP").count()
```

```python
# remove malformed age groups
malformed = ["14-Oct", "4-Jan", "9-May"]
ddf4 = ddf3[~ddf3["AGE GROUP"].isin(malformed)]
ddf4.groupby("AGE GROUP").count()
```

```python
ddf4.groupby("CAUSE OF DEATH").count()
```

```python
ddf4.groupby("PLACE OF DEATH").count()
```

```python
# remove not classifiable place of deaths
ddf5 = ddf4[ddf4["PLACE OF DEATH"] != "Not Classifiable"]
ddf5.groupby("PLACE OF DEATH").count()
```

```python
ddf5.count()
```

```python
deaths_df.count()
```

```python
import numpy as np

rows = ddf5.shape[0]

train_rows = round(rows * 0.8)

train, test = np.split(ddf5, [train_rows])

train.count()
```

```python
# xs_columns = ["COUNTY OF RESIDENCE", "GENDER", "RACE", "EDUCATION"]
# ys_columns = ["AGE GROUP", "CAUSE OF DEATH", "PLACE OF DEATH"]
xs_columns = ["GENDER", "EDUCATION"]
ys_columns = ["AGE GROUP"]

x_train = train[xs_columns]
y_train = train[ys_columns]

x_test = test[xs_columns]
y_test = test[ys_columns]
```

```python
xs_meta = {}

# create model key for exporting to json
for column in x_train.columns:
    xs_meta[column] = list(x_train[column].unique())
    
xs_meta
```

```python
ys_meta = {}

# create model key for exporting to json
for column in y_train.columns:
    ys_meta[column] = list(y_train[column].unique())
    
ys_meta
```

```python
def one_hot_column(df, column):
    one_hot = pd.get_dummies(df[column])
    return df.copy().drop(column, axis=1).join(one_hot)

def one_hot_df(df):
    new_df = df.copy()
    
    for column in list(df.columns):
        new_df = one_hot_column(new_df, column)
        
    return new_df
    

xoh_train = one_hot_df(x_train)
yoh_train = one_hot_df(y_train)

xoh_train.head()
```

```python
from sklearn import linear_model, metrics

# ordinary least squares with multivariate linear regression
reg = linear_model.LinearRegression()
reg.fit(xoh_train, yoh_train)
```

```python
xoh_test = one_hot_df(x_test)
yoh_test = one_hot_df(y_test)

y_pred = reg.predict(xoh_test)
```

```python
print("Coefficients: ", reg.coef_)
print("Intercepts: ", reg.intercept_)
```

```python
mse = metrics.mean_squared_error(yoh_test, y_pred)
print("Mean Squared Error: %0.2f" % mse)
```

```python
r2 = metrics.r2_score(yoh_test, y_pred)
print("Variance Score: %0.2f" % r2)
```
